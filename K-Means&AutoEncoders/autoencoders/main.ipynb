**K means Clustering - Defining Model**

```python
#Defining the K-means functions to calculate the centroids
def kmeans(inputparam,k, no_of_iterations):
    randomcent = np.random.choice(len(inputparam), k, replace=False)
    #Randomly choosing Centroids 
    centroids = inputparam[randomcent, :]
    #finding the distance between centroids and all the data points
    distances = cdist(inputparam, centroids ,'euclidean')
    #Centroid with the minimum Distance
    points = np.array([np.argmin(i) for i in distances]) 
    #Repeating the above steps for a defined number of iterations
    for _ in range(no_of_iterations): 
        centroids = []
        for randomcent in range(k):
            #Updating Centroids by taking mean of Cluster it belongs to
            centroids.append(inputparam[points==randomcent].mean(axis=0))

        centroids = np.vstack(centroids) #Updated Centroids according to distance
        distances = cdist(inputparam, centroids ,'euclidean')
        points = np.array([np.argmin(i) for i in distances])
         
    return points 
```
